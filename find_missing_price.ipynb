```python
import pandas as pd
from os import path, listdir
```

```python
price_dir = "./data/daily_price_sp500"
master_list = "./data/s&p500_historical_companies_1990_expanded.csv"
```

```python
data_status = []
for f in listdir(price_dir):
    ticker = f.split('_')[0]
    figi = f.split('_')[1]
    entry = {'Ticker': ticker, 'figi': figi}
    if path.exists(path.join(price_dir, f)):
        entry['has_file'] = 1
        try:
            df = pd.read_csv(path.join(price_dir, f))
            entry['has_data'] = 1
        except Exception as e:
            entry['has_data'] = 0
            pass
    else:
        entry['has_file'] = 0
    data_status.append(entry)
```

```python
df_status = pd.DataFrame(data_status)
```

```python
df_status
```

```python
df_master = pd.read_csv(master_list)
```

```python
df_master
```

```python
df_master = pd.merge(df_master, df_status, on=['Ticker', 'figi'], how='left')
```

```python
df_master
```

```python
df_master.fillna({'has_file': 0, 'has_data': 0}, inplace=True)
```

```python
df_master.to_csv('s&p500_historical_company_1990_price_status.csv', index=False)
```
